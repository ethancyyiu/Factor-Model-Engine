```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from factor_engines.load_data import get_price_data, compute_returns
from factor_engines.factors import (
    momentum_factor,
    volatility_factor,
    reversal_factor,
    sma_distance_factor,
    compute_all,
    zscore,
    daily_score
)
from factor_engines.portfolio import (
    rank_factor,
    long_short_weights,
    normalize,
    portfolio_returns,
    zscore_weights,
    zscore_all_factors,
    composite_factor,
    build_long_short_portfolio,
    build_zscore_portfolio_from_factor,
    build_composite_portfolio
)

plt.style.use("seaborn-v0_8")
```

```python
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "META"]

prices = get_price_data(tickers, "2015-01-01", "2025-01-01")
returns = compute_returns(prices)

prices.head()
returns.head()
```

```python
momentum = momentum_factor(prices)
volatility = volatility_factor(prices)
reversal = reversal_factor(prices)
sma = sma_distance_factor(prices)

factors = {
    "momentum": momentum,
    "volatility": volatility,
    "reversal": reversal,
    "sma_distance": sma
}
```

```python
mom_weights, mom_ret = build_long_short_portfolio(momentum, returns, long_pct=0.2, short_pct=0.2)
mom_ret.head()
```

```python
mom_weights_z, mom_ret_z = build_zscore_portfolio_from_factor(momentum, returns)
mom_ret_z.head()
```

```python
comp_weights, comp_ret = build_composite_portfolio(factors, returns)
comp_ret.head()
```

```python
def cumulative_returns(returns):
    return (1 + returns).cumprod()

cum_mom_ls = cumulative_returns(mom_ret)
cum_mom_z = cumulative_returns(mom_ret_z)
cum_comp = cumulative_returns(comp_ret)

plt.figure(figsize=(12, 6))
plt.plot(cum_mom_ls, label="Momentum Long/Short")
plt.plot(cum_mom_z, label="Momentum Z-Score")
plt.plot(cum_comp, label="Composite Z-Score")
plt.title("Cumulative Returns of Factor Portfolios")
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend()
plt.grid(True)
plt.show()

```

```python
def sharpe_ratio(returns_series, periods_per_year=252):
    mean_ret = returns_series.mean()
    vol = returns_series.std()
    if vol == 0:
        return np.nam
    else:
        daily_sharpe = mean_ret / vol 
    return daily_sharpe * np.sqrt(periods_per_year)

strategies = {
    "Momentum Long/Short": mom_ret,
    "Momentum Z-Score": mom_ret_z,
    "Composite Z-Score": comp_ret
}

for name, ret in strategies.items():
    sr = sharpe_ratio(ret)
    print(f"{name}: Sharpe = {sr:.2f}")

```

```python
vol_weights_ls, vol_ret_ls = build_long_short_portfolio(volatility, returns)
rev_weights_ls, rev_ret_ls = build_long_short_portfolio(reversal, returns)
sma_weights_ls, sma_ret_ls = build_long_short_portfolio(sma, returns)

summary = []

for name, ret in {
    "Momentum LS": mom_ret,
    "Momentum Z": mom_ret_z,
    "Composite Z": comp_ret,
}.items():
    sr = sharpe_ratio(ret)
    ann_ret = (1 + ret).prod() ** (252 / len(ret)) - 1
    ann_vol = ret.std() * np.sqrt(252)
    summary.append([name, ann_ret, ann_vol, sr])

summary_df = pd.DataFrame(summary, columns=["Strategy", "Ann Return", "Ann Vol", "Sharpe"])
summary_df
```
